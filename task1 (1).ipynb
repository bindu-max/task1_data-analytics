```python

# TASK 1: DATA IMMERSION & WRANGLING
# Dataset: ApexPlanet_DataAnalytics_Dataset.xlsx
import pandas as pd

# STEP 1: LOAD DATASET
df = pd.read_excel("ApexPlanet_DataAnalytics_Dataset (1).xlsx")

print("Original Shape:", df.shape)

# STEP 2: DATA EXPLORATION
print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Records:", df.duplicated().sum())


# STEP 3: HANDLE MISSING VALUES
# Fill Age with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill City with mode
df["City"] = df["City"].fillna(df["City"].mode()[0])

# STEP 4: REMOVE DUPLICATES
df = df.drop_duplicates()


# STEP 5: CONVERT DATA TYPES
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
df["Age"] = df["Age"].astype(int)
df["Quantity"] = df["Quantity"].astype(int)
df["Unit_Price"] = df["Unit_Price"].astype(float)
df["Total_Sales"] = df["Total_Sales"].astype(float)
# STEP 6: RENAME COLUMNS
df.rename(columns={
    "Order_ID": "OrderID",
    "Order_Date": "OrderDate",
    "Customer_ID": "CustomerID",
    "Customer_Name": "CustomerName",
    "Unit_Price": "UnitPrice",
    "Total_Sales": "TotalSales"
},inplace=True)

# STEP 7: FEATURE ENGINEERING
df["Year"] = df["OrderDate"].dt.year
df["Month"] = df["OrderDate"].dt.month
df["Day"] = df["OrderDate"].dt.day

# STEP 8: FINAL DATA CHECK
print("\nMissing Values After handling:")
print(df.isnull().sum())
print("\nCleaned Dataset Shape:", df.shape)

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

print("\nData Types:")
print(df.dtypes)

# STEP 9: SAVE CLEANED DATASET
df.to_excel("Cleaned_Sales_Dataset.xlsx", index=False)

print("\nCleaning Completed Successfully!")
print("Cleaned dataset saved as: Cleaned_Sales_Dataset.xlsx")
```
